```python
display(students_frame['國文']) #有索引index
for item in students_frame['國文']:
    print(item)

#只能得到value,沒有索引index
```

```python
for item in students_frame['國文']:
    if item >= 80:
        print("優")
    elif item >= 70:
        print("甲")
    elif item >= 60:
        print("乙")
    else:
        print("丙")
```

```python
def func1(item):
    if item >= 80:
        return'優'
    elif item >= 70:
        return'甲'
    elif item >= 60:
        return'乙'
    else:
        return'丙'
    
display(students_frame['國文'].map(func1))
```

### 新增欄

```python
students_frame['國文等級'] = students_frame['國文'].map(func1)
students_frame
```

### DataFrame.apply()

```python
students_frame
```

```python
def func2(s):
    return s.min()

students_frame.apply(func2)
```

```python
students_frame.apply(func2,axis = 1)
```

```python
def func3(s):
    return s [s < 60].count()
students_frame.apply(func3,axis = 1)
#不及格的科目數量
```

```python
students_frame['不及格科目數'] = students_frame.apply(func3,axis = 1)
students_frame
```

```python
students_frame = frame[["國文","英文","數學","地理","歷史"]]
students_frame
```

```python
def func5(s):
    return pd.Series([s.max(), s.min(), s.median(), s.mean(), s.std()],
                     index = ["最大值", "最小值", "中位數","平均值","標準差"])
students_frame.apply(func5)
```

### 相關係數

```python
import numpy as np
import pandas as pd

data = {'年廣告費投入':[12.5, 15.3, 23.2, 26.4, 33.5, 34.4, 39.4, 45.2, 55.4, 60.9],
        '月均銷售額':[21.2, 23.9, 32.9, 34.1, 42.5, 43.2, 49.0, 52.8, 59.4, 63.5]}
dataframe = pd.DataFrame(data, index=[2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020])
dataframe['年廣告費投入'].corr(dataframe['月均銷售額'])
```

```python
import matplotlib.pyplot as plt

figure = plt.figure(figsize = (8, 5))
axes = figure.add_subplot(1,1,1)
axes.plot(dataframe.index,dataframe['年廣告費投入'], 'r-o')
axes.plot(dataframe.index,dataframe['月均銷售額'], 'b-o')
plt.show()
```
