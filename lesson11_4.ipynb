```python
#讀取csv二維list
import csv

with open('a2.csv', encoding ='utf-8', newline= '' ) as file:
    csv_reader = csv.reader(file)
    next(csv_reader)
    #print(type(csv_reader))
    csv_list = list(csv_reader)
display(csv_list)   
```

```python
for student in csv_list:
    if student[0] == '周家銘':
        print(student)
        break
```
